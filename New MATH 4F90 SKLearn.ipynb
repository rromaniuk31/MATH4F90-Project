```python
preddf = preddf.groupby('GameID').agg(lambda x: x.tolist())
preddf[['Team','Opponent']] = pd.DataFrame(preddf.Team.tolist(), index= preddf.index)
preddf[['Actual_PF','Opp Actual PF']] = pd.DataFrame(preddf.Actual_PF.tolist(), index= preddf.index)
preddf[['Predicted_PF','Opp Pred PF']] = pd.DataFrame(preddf.Predicted_PF.tolist(), index= preddf.index)
preddf.to_csv(r"D:\\Fourth Year\\MATH 4F90\\fullpredictions.csv", index = False, sep = ',', encoding = 'utf-8')
preddf
```

```python
preddf['Actual Team1 Won'] = ''
preddf['Predicted Team1 Won'] = ''
preddf['Prediction Correct?'] = ''

for i in range(1,135):
    if preddf['Actual_PF'][i] > preddf['Opp Actual PF'][i]:
        preddf['Actual Team1 Won'][i] = True
    else:
        preddf['Actual Team1 Won'][i] = False
        
for i in range(1,135):
    if preddf['Predicted_PF'][i] > preddf['Opp Pred PF'][i]:
        preddf['Predicted Team1 Won'][i] = True
    else:
        preddf['Predicted Team1 Won'][i] = False

for i in range(1,135):
    if preddf['Actual Team1 Won'][i] == preddf['Predicted Team1 Won'][i]:
        preddf['Prediction Correct?'][i] = 1
    else:
        preddf['Prediction Correct?'][i] = 0

preddf
```

```python
correct_preds = preddf['Prediction Correct?'].sum()
total_preds = preddf['Prediction Correct?'].count()
prop_correct = correct_preds/total_preds
print("{}\n".format(correct_preds), "{}\n".format(total_preds), "{}\n".format(prop_correct))
```

## Calculate MSE

```python
MSEdf = pd.DataFrame(data=ypred, columns = ['Predictions'])
MSEdf['Actual'] = ''
MSEdf['Team'] = ''

for i in range(0,268):
    MSEdf['Actual'][i] = df['PF**'][i]
    MSEdf['Team'][i] = df['Team'][i]

MSEdf['y-yhat'] = ''
MSEdf['y-yhat_sq'] = ''

for i in range(0,268):
    MSEdf['y-yhat'][i] = MSEdf['Actual'][i] - MSEdf['Predictions'][i]
    MSEdf['y-yhat_sq'][i] = MSEdf['y-yhat'][i]**2

MSEdf
```

```python
MSE = MSEdf['y-yhat_sq'].sum()/MSEdf['y-yhat_sq'].count()
MSE
```
